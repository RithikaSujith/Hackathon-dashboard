Importing the libraries

```
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(rc={'figure.figsize':(15.7,8.27)})
```

Importing the dataset

```
df=pd.read_excel('state_wise_updated.xlsx')
df.head()
```

```
df1=df.iloc[1:]
df1.head()
```

```
df1['50-60']=df1['50-54']+df1['55-59']
df1['60-70']=df1['60-64']+df1['65-69']
df1['70-80']=df1['70-74']+df1['75-79']
```

```
df1.head()
```

```
df_50_60 =df1[['State','50-60','Vaccination Centres']]
df_60_70 =df1[['State','60-70','Vaccination Centres']]
df_70_80 =df1[['State','70-80','Vaccination Centres']]
df_80 =df1[['State','80+','Vaccination Centres']]
```

```
df_50_60.head()
```

```
df_60_70.head()
```

```
df_70_80.head()
```

```
df_80.head()
```

```
df_50_60['per_day_vaccination'] =df_50_60['Vaccination Centres']*100
df_60_70['per_day_vaccination'] =df_60_70['Vaccination Centres']*100
df_70_80['per_day_vaccination'] =df_70_80['Vaccination Centres']*100
```

```
df_80['per_day_vaccination'] =df_80['Vaccination Centres']*100
```

```
df_50_60.head()
```

```
df_50_60['days_needed'] =df_50_60['50-60'] / df_50_60['per_day_vaccination']
df_60_70['days_needed'] =df_60_70['60-70'] / df_60_70['per_day_vaccination']
df_70_80['days_needed'] =df_70_80['70-80'] / df_70_80['per_day_vaccination']
df_80['days_needed'] =df_80['80+']/df_80['per_day_vaccination']
```

```
df_50_index=pd.read_json('50plus_index (2)')
df_60_index=pd.read_json("60plus_index (1)")
df_70_index = pd.read_json("70plus_index (3)")
df_80_index=pd.read_json('80plus_index (1)')
```

```
df_50_index.head(30)
```

```
df_50_60.set_index('State' , inplace=True)
df_60_70.set_index('State', inplace=True)
df_70_80.set_index('State', inplace=True)
df_80.set_index('State', inplace=True)
```

```
df_50_60=df_50_60.join(df_50_index)
df_60_70=df_60_70.join(df_60_index)
df_70_80=df_70_80.join(df_70_index)
df_80=df_80.join(df_80_index)
df_50_60.head(30)
```

```
df_50_60.reset_index(inplace=True)
df_60_70.reset_index(inplace=True)
df_70_80.reset_index(inplace=True)
df_80.reset_index(inplace=True)
```

```
df_80.groupby('index').max()
```

```
df_50_60.to_json('phase2_state_50_60',orient='records')
df_60_70.to_json('phase2_state_60_70',orient='records')
df_70_80.to_json('phase2_state_70_80',orient='records')
df_80.to_json('phase2_state_80',orient='records')
```
